```python
print("Missing Values in Each Column:")
print(df.isnull().sum())
```

```python
print("Number of Duplicate Rows:")
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print("Shape of Dataset after Removing Duplicates:")
print(df.shape)
```

```python
print("Statistical Summary:")                   # Statistical summary of all numeric columns
print(df.describe())
```

```python
print("---------Salary Statistics------")                      # Salary Statistics
print("Minimum Salary:", df["Salary"].min())
print("Maximum Salary:", df["Salary"].max())
print("Mean Salary:", df["Salary"].mean())
print("Median Salary:", df["Salary"].median())

print("\n-------Age Statistics--------")                       # Age Statistics
print("Minimum Age:", df["Age"].min())
print("Maximum Age:", df["Age"].max())
print("Mean Age:", df["Age"].mean())
print("Median Age:", df["Age"].median())

print("\n--------Years of Experience Statistics-------")                   # Years of Experience Statistics
print("Minimum Experience:", df["Years of Experience"].min())
print("Maximum Experience:", df["Years of Experience"].max())
print("Mean Experience:", df["Years of Experience"].mean())
print("Median Experience:", df["Years of Experience"].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(df["Age"], bins=20, edgecolor="black")           # Histogram for Age
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
plt.hist(df["Years of Experience"], bins=20, edgecolor="black")           # Histogram for Years of Experience
plt.title("Histogram of Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Frequency")
```

```python
plt.hist(df["Salary"], bins=20, edgecolor="black")          # Histogram for Salary
plt.title("Histogram of Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x="Gender", data=df)            # Count Plot for Gender
plt.title("Count Plot of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x="Education Level", data=df)             # Count Plot for Education Level
plt.title("Count Plot of Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
```

```python
sns.countplot(x="Job Title", data=df)                   # Count Plot for Job Title
plt.title("Count Plot of Job Title")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
```

```python
sns.scatterplot(x='Years of Experience', y='Salary', data=df)
plt.title("Years of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
sns.regplot(x='Years of Experience', y='Salary', data=df)

plt.title("Years of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
sns.boxplot(y='Salary', data=df)

plt.title("Box Plot of Salary")
plt.ylabel("Salary")
```

```python
sns.barplot(x='Education Level', y='Salary', data=df)
plt.title("Average Salary by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Average Salary")
```
