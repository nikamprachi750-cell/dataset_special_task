```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['Sales'])                     # Histogram of Sales
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Count")
```

```python
sns.countplot(x='Category',data=df)                       # Countplot of Category
plt.title("Count of Products by Category")
plt.show()
```

```python
sns.heatmap(df[num_cols].corr(),annot=True,linewidths=0.5)        # Correlation Heatmap

plt.title("Correlation Heatmap")
```

```python
sns.boxplot(x=df["Sales"])                        # Boxplot of Sales

plt.title("Boxplot of Sales")
plt.xlabel("Sales")

plt.show()
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')                  # Total Sales by Region
                                                                      #bar chart
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
df["Category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Category Distribution")
plt.show()
```
